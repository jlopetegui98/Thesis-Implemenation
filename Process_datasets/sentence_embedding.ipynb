```python
import multiprocessing
from utils import get_embeddings
import torch
from joblib import Parallel, delayed
from joblib._memmapping_reducer import has_shareable_memory
from transformers import BertTokenizer, BertModel
import time

# Load pre-trained model tokenizer (vocabulary)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Datasets path
DATA_PATH = "content/drive/My Drive/datasets/"
```

## Expertise dataset

```python
with open(DATA_PATH + "expertise/sentences/train.txt", "r") as f:
    sentences = f.readlines()

embeddings_list = [None for _ in range(len(sentences))]

# get_embeddings(sentences,embeddings_list,tokenizer,0,1)
Parallel(n_jobs=4, backend="multiprocessing")(delayed(has_shareable_memory)(get_embeddings(sentences,embeddings_list,tokenizer,i,i+5)) for i in range(0,len(sentences),1000))
```

```python
torch.save(embeddings_list, DATA_PATH + "embeddings_dataset/expertise/train.pt")
```

```python
with open(DATA_PATH + "expertise/sentences/test.txt", "r") as f:
    sentences_test = f.readlines()

embeddings_list_test = [None for _ in range(len(sentences_test))]

# get_embeddings(sentences,embeddings_list,tokenizer,0,1)
Parallel(n_jobs=4, backend="multiprocessing")(delayed(has_shareable_memory)(get_embeddings(sentences_test,embeddings_list_test,tokenizer,i,i+5)) for i in range(0,len(sentences_test),1000))
```

```python
torch.save(embeddings_list_test, DATA_PATH + "embeddings_dataset/expertise/test.pt")
```
